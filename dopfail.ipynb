# Анализируем данные

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")  # белая сетка
sns.set_palette("pastel")  # нежные цвета

p = Path("data/xlsx/")
for x in p.rglob("*"):
    df = pd.read_excel(x, index_col=0)
    
    # Создание графика
    plt.figure(figsize=(12, 6))  # размер графика
    sns.lineplot(x='dt', y=df.columns[-1], data=df, marker='o', linewidth=2.5)
    
    # Настройки заголовка и меток
    plt.title(str(x)[10:-5], fontsize=16, fontweight='bold')
    plt.xlabel('Дата', fontsize=14)
    plt.ylabel(df.columns[-1], fontsize=14)
    plt.xticks(rotation=45)  # поворот меток по оси X
    plt.tight_layout()  # улучшение размещения элементов на графике
    
    plt.show()
```

```python
df = pd.read_excel("data/train.xlsx")
```

```python
sns.set(style="whitegrid")
sns.set_palette("pink")  # розовая палитра

# Предположим, что df уже загружен ранее
plt.figure(figsize=(14, 6))

# Добавляем график
sns.lineplot(data=df, x=df.index, y='Цена на арматуру', marker='o', linewidth=2.5)

# Настройки заголовка и меток
plt.title('Цена на арматуру', fontsize=18, fontweight='bold', color='darkmagenta')
plt.xlabel('Дата (dt)', fontsize=15)
plt.ylabel('Цена на арматуру', rotation=70, labelpad=30, fontsize=15)
plt.xticks(rotation=45)
plt.tight_layout()

# Отображение графика
plt.show()
```
